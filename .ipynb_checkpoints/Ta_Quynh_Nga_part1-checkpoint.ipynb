```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
from tqdm.notebook import tqdm
```

There are 82 members in SCSE faculty and the result is spread out into 2 DR-NTU pages. Therefore, I search through these 2 pages (2 urls) to retrieve all 82-member information, using `BeautifulSoup`

```python
dr_pages = [
    "https://dr.ntu.edu.sg/simple-search?filterquery=ou00030&filtername=school&filtertype=authority&location=researcherprofiles",
    "https://dr.ntu.edu.sg/simple-search?query=&location=researcherprofiles&filter_field_1=school&filter_type_1=authority&filter_value_1=ou00030&crisID=&relationName=&sort_by=score&order=desc&rpp=50&etal=0&start=50",
]

df = pd.DataFrame(columns=['Full Name', 'Email', 'DR-NTU URL'])

for url in dr_pages:
    soup_source = requests.get(url).text
    soup = BeautifulSoup(soup_source,'lxml') 
    for row in soup.find("table", class_="table table-hover").find_all('tr')[1:]:
        name = row.find('td', headers='t1').text
        dr_url = 'https://dr.ntu.edu.sg' + row.find('td', headers='t1').a['href']
        email = row.find('td', headers='t3').text
        df = df.append({'Full Name': name,
                        'Email': email,
                        'DR-NTU URL': dr_url}, ignore_index=True)
df
```

Then sort the DataFrame by `Email`

```python
df = df.sort_values(by='Email', ignore_index=True)
df
```

```python
df['Website URL'] = np.nan
df['DBLP URL'] = np.nan
```

The value `Website URL` are retrieved by accessing the DR-URL individually and then searching for the website url using `BeautifulSoup` in `href` of the tag `div` that has `id=personalsiteDiv`

```python
def get_website_URL(dr_url):
    soup_source = requests.get(dr_url).text
    soup = BeautifulSoup(soup_source,'lxml')
    try:
        website_URL = soup.find('div', id='personalsiteDiv').a['href']
    except:
        if soup.find('div', id='personalsiteDiv') == None:
            return np.nan
    return website_URL
```

```python
df['Website URL'] = df['DR-NTU URL'].apply(get_website_URL)
```

To get the `DBLP URL`, we need to search with the url format https://dblp.org/search/author?q={Full_Name}. The `Full Name` needs to be preprocessed to remove chinese characters

Since the page is dynamic (can change its url and interface after we search the aforementioned url), the html body can have class as `no-js search-page` or `no-js pers-page`. 
- In `no-js pers-page`, the page goes directly to the member's DBLP url, and its url can be found in `soup.find('body', class_='no-js pers-page')['data-dblp-path']`.
- In `no-js search-page`, the page shows a list of people that matches the `Full Name`, so we need to go to each list item to check if person comes from NTU or not. If yes, extract the DBLP URL of that list item

```python
def get_DBLP_URL_naive(name):
    name = re.sub('\（.+\）', '', name).strip()
    search_url = f'https://dblp.org/search/author?q={name}'
    soup_source = requests.get(search_url).text
    soup = BeautifulSoup(soup_source,'lxml')
    
    if soup.find('body', class_='no-js search-page') != None:
        em = soup.find('div', id='completesearch-authors').find('div', class_='body hide-body').em
        if em != None and em.text == 'no matches':
            print(f'{name} NO MATCHES')
            return np.nan
        
        all_matches = soup.find_all('ul', class_='result-list')[-1]
        found = False
        for li in all_matches:
            if 'Nanyang Technological University' in li.text and 'Computer' in li.text:
                print(f'{name} seach_page')
                found = True
                return li.a['href'] + '.html'
        if not found:
            for li in all_matches:
                if 'Nanyang Technological University' in li.text:
                    print(f'{name} seach_page')
                    found = True
                    return li.a['href'] + '.html'
    
    if soup.find('body', class_='no-js pers-page') != None:
        dblp_path = soup.find('body', class_='no-js pers-page')['data-dblp-path'].replace('pers', 'pid')
        print(f'{name} pers_page')
        return f'https://dblp.org/{dblp_path}.html'
        
    print(f'{name} \t NaN')
    return np.nan
```

Note here: I used `df.iterrows()` to process each row one by one instead of using the `pandas.DataFrame.apply()` function. This is because `apply()` is notorious for running slow in many cases, especially when the function inside `apply()` is complicated

```python
for i, row in tqdm(df.iterrows(), total=len(df)):
    try:
        df.at[i, 'DBLP URL'] = get_DBLP_URL_naive(row['Full Name'])
    except:
        print(f"ERROR AT {i} {row['Full Name']}")
```

To get the number of `Citations (All)`, I used `BeautifulSoup` to access through https://scholar.google.com/citations?view_op=search_authors&hl=en&mauthors={Full_Name} and search through all person appearing in the response, then I checked if the person comes from NTU or not to decide if we should use this profile information here. 

Since the response can spread into many pages (each page 10 people), we need to add another parameter variable `astart` to retrieve the next 10 responses. However, sometimes, the response has so many pages (hundreds of pages), I set a `limit=100` so that I just search through first 100 responses (100 people)

```python
df['Citations (All)'] = 0

def get_citations(name, astart=0, limit=100):
    if astart > limit:
        return np.nan
    name = re.sub('\（.+\）', '', name).strip()
    search_url = f'https://scholar.google.com/citations?view_op=search_authors&hl=en&mauthors={name}&astart={astart}'
    soup_source = requests.get(search_url).text
    soup = BeautifulSoup(soup_source,'lxml')
    
    list_search = soup.find_all('div', class_='gs_ai gs_scl gs_ai_chpr')
    for person in list_search:
        print(person.text)
        if person.find('div', class_='gs_ai_eml').text == 'Verified email at ntu.edu.sg' or\
           'Nanyang Technological University' in person.find('div', class_='gs_ai_aff'): 
            citations = person.find('div', class_='gs_ai_cby').text
            return re.findall(r'\d+', citations)[0]
        
    if soup.find('div', id='gsc_authors_bottom_pag') != None:
        page_range = soup.find('div', id='gsc_authors_bottom_pag').text
        astart = int(page_range.split(' - ')[1])
        if int(astart) % 10 == 0:
#             print(f'* * * * * {name} NEXT PAGE {astart}')
            return get_citations(name, astart=astart)
    return np.nan
```

I search for `Citations` for the 2nd round. 

Sometimes, searching on Google Scholar by another names instead of the `Full Name` works, such as the name from DBLP page. Therefore, in this 2nd round, I search the "another name" in DBLP first (which mostly as same as the name in their research paper citation --> easier to search in Google Scholar), then use that name to search in Google Scholar.

If still cannot find the profile in the 2nd round, search for the 3rd round with additional keyword `nanyang`

```python
def get_citations_1(dblp_url):
    soup_source = requests.get(dblp_url).text
    soup = BeautifulSoup(soup_source,'lxml')
    
    name = soup.find('header', id='headline').find('span', class_='name primary').text
    print(name)
    return get_citations_(name), name
```

```python
for i, row in tqdm(df[df['Citations (All)'].isna()].iterrows(), total=len(df[df['Citations (All)'].isna()])):
    if not pd.isna(row['DBLP URL']):
        soup_source = requests.get(row['DBLP URL']).text
        soup = BeautifulSoup(soup_source,'lxml')
        name = soup.find('header', id='headline').find('span', class_='name primary').text
        print(f'{i} {name}')
        df.at[i, 'Citations (All)'], name = get_citations_1(row['DBLP URL'])
        
    # if still cannot find the profile in the 2nd round, search for the 3rd round with additional keyword 'nanyang'
    if pd.isna(row['Citations (All)']):
        df.at[i, 'Citations (All)'] = get_citations_(name + 'nanyang')
```
